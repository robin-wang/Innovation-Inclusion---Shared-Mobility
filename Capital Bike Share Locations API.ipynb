### ESTABLISH API CONNECTION

```python
import requests
import json
from IPython.display import display_javascript, display_html, display
```

```python
response = requests.get("https://opendata.arcgis.com/datasets/a1f7acf65795451d89f0a38565a975b3_5.geojson")
```

```python
data = response.json()
print(type(data))
print(data)
parsed = json.loads(data)
print(type(parsed))
```

```python
#data structure check
parsed['features'][0]['properties']
```

```python
objectid, ID, Address, Terminal_Number, lat, lon, Installed, locked, install_date,removal_date,temporary_install,number_bike,number_empty,x,y,se_anno_cad_data,owner = [],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[]
for result in parsed['features']:
    objectid.append(result['properties']['OBJECTID'])
    ID.append(result['properties']['ID'])
    Address.append(result['properties']['ADDRESS'])
    Terminal_Number.append(result['properties']['TERMINAL_NUMBER'])
    lat.append(result['properties']['LATITUDE'])
    lon.append(result['properties']['LONGITUDE'])
    Installed.append(result['properties']['INSTALLED'])
    locked.append(result['properties']['LOCKED'])
    install_date.append(result['properties']['INSTALL_DATE'])
    removal_date.append(result['properties']['REMOVAL_DATE'])
    temporary_install.append(result['properties']['TEMPORARY_INSTALL'])
    number_bike.append(result['properties']['NUMBER_OF_BIKES'])
    number_empty.append(result['properties']['NUMBER_OF_EMPTY_DOCKS'])
    x.append(result['properties']['X'])
    y.append(result['properties']['Y'])
    se_anno_cad_data.append(result['properties']['SE_ANNO_CAD_DATA'])
    owner.append(result['properties']['OWNER'])

df_opendata = pd.DataFrame([objectid, ID, Address, Terminal_Number, lat, lon, Installed, locked, install_date,removal_date,temporary_install,number_bike,number_empty,x,y,se_anno_cad_data,owner],
                          index = ['objectid','ID','Address','Terminal_Number','lat','lon','Installed','locked','install_date',
                                   'removal_date','temporary_install','number_bike','number_empty','x','y','se_anno_cad_data',
                                   'owner']).T
df_opendata.shape
```
